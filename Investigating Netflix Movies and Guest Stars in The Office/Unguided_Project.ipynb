```python
# Use this cell to begin your analysis, and add as many as you would like!
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# function to specify a color scheme reflecting the scaled ratings
def color_(x):
    if x < 0.25:
        return "red"
    elif x < 0.50:
        return "orange"
    elif x < 0.75:
        return "lightgreen"
    else:
        return "darkgreen"
```

```python
# function to specify the size by guest appearances
def size_(x):
    if x == True :
        return "red"
    elif x < 0.50:
        return "orange"
    elif x < 0.75:
        return "lightgreen"
    else:
        return "darkgreen"
```

```python
# data loading and creation of new features
df = pd.read_csv("datasets/office_episodes.csv")
df['rank_c'] = df['scaled_ratings'].apply(color_)
df['rank_S'] = df['has_guests'].apply(lambda x: 250 if x == True else 25)
df.head()
```

```python
# separation of dataframes if there was a guest or not in the episode
df_WS = df[df['has_guests'] == True]
df_NS = df[df['has_guests'] == False]
```

```python
# plot figure
fig = plt.figure(figsize=(12,5))
plt.scatter(df_WS['episode_number'], df_WS['viewership_mil'], c=df_WS['rank_c'], s=df_WS['rank_S'], marker="*")
plt.scatter(df_NS['episode_number'], df_NS['viewership_mil'], c=df_NS['rank_c'], s=df_NS['rank_S'], marker="o")
plt.ylabel('Viewership (Millions)')
plt.xlabel('Episode Number')
plt.title("Popularity, Quality, and Guest Appearances on the Office")
plt.show()
```
